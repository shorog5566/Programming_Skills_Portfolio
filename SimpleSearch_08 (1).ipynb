```python
names = ["Jake", "Zac", "Ian", "Ron", "Shoroog", "Dave"]    # list of names

if "Shoroog" in names:
    print("Shoroog found!")
else:
    print("Shoroog not found!")
```

```python
names = ["Jake", "Zac", "Ian", "Ron", "Shoroog", "Dave"]

search_name = input("Enter the name to search: ")     # user input

if search_name in names:
    print(search_name, "found!")
else:
    print(search_name, "not found!")
```
