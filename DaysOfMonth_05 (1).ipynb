```python
months = {
    1: 31,
    2: 28,
    3: 31,
    4: 30,
    5: 31,
    6: 30,
    7: 31,
    8: 31,
    9: 30,
    10: 31,
    11: 30,
    12: 31
}

month = int(input("Enter the month number (1-10): "))

if month == 2:                        # check february
    leap = input("Is it a leap year? (yes/no): ")
    if leap.lower() == "yes":
        print(29)              # leap year days
    else:
        print(28)         # normal days
elif month in months:
    print(months[month])
else:
    print("Invalid month number.")
```
