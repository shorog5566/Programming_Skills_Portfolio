```python
def is_even(number):        # function definition
    if number % 2 == 0:
        return "Even"       # even number
    else:
        return "Odd"        # odd number

def main():
    num = int(input("Enter a number: "))     # user input
    result = is_even(num)     # check number 
    print(result)

if __name__ == "__main__":
    main()
```
